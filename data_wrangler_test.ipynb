```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
titanic_url = 'https://raw.githubusercontent.com/MarkWilcock/CourseDatasets/main/Misc%20Datasets/Titanic.csv'
df = pd.read_csv(titanic_url) # read the data
```

```python
# needs to be in own cell to launcg DW
df.head()
```

```python
"""
Cell generated by Data Wrangler - follows the video tutorial
"""
def clean_data(df):
    # Drop column: 'Cabin'
    df.drop(columns=['Cabin'], inplace=True)

    # Replace missing values with the mean of each column in: 'Age'
    df.fillna({'Age': int(df['Age'].mean())}, inplace=True)

    df.insert(4, "FirstName", df["Name"].str.split(" ").str[2])

    # Derive column 'FamilySize' from columns: 'SibSp', 'Parch'
    def FamilySize(Parch, SibSp):
        """
        Transform "Parch", "SibSp" as per the following examples:
          0, 0 ==> 0
          1, 3 ==> 4
          0, 1 ==> 1
        """
        number1 = Parch + SibSp + 1
        return f"{number1:01.0f}"
    
    df.insert(9, "FamilySize", df.apply(lambda row : FamilySize(row["Parch"], row["SibSp"]), axis=1))

    df.insert(6, "SexNumber", df['Sex'].apply(lambda x: 1 if x == 'male' else 0))
    # Replace missing values with the mean of each column in: 'Age'
    df = df.loc[:, ['Survived', 'Pclass', 'SexNumber', 'Age', 'FamilySize', 'Embarked']]
    df = pd.get_dummies(df, columns=['Embarked'])
    return df

df_clean = clean_data(df.copy())
df_clean.head()
```

```python
X = df_clean.drop('Survived', axis=1)
y = df_clean['Survived']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
(X_train.shape, y_train.shape), (X_test.shape, y_test.shape)
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
predictions = model.predict(X_test)
accuracy_score(y_test, predictions)
```
