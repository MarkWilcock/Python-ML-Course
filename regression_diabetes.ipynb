```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets, linear_model, model_selection
```

The diabetes dataset is an internal example dataset in the sklean package

```python
X, y = datasets.load_diabetes(return_X_y=True)
print(X.shape)
print(X[0]) # first row
print(X[:5, 3]) # first 5 values of the 4th column
```

```python
X = X[:, 2]  # get all rows, only 3rd column from the dataset, the BMI value
X.shape
```

```python
#  Create a 2D array from the 1D array by reshaping it to (-1,1)
X = X.reshape((-1,1))
```

```python
# spilt into train and test datasets
X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size=0.33)
```

```python
# Creeate a linear regression model and train it with the training data set 
model = linear_model.LinearRegression()
model.fit(X_train, y_train)
```

```python
# create a prediction using the test data set
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
plt.scatter(X_test, y_test,  color='black')
plt.scatter(X_test, y_pred,  color='green')
#plt.plot(X_test, y_pred, color='blue', linewidth=3)
plt.xlabel('Scaled BMIs')
plt.ylabel('Disease Progression')
plt.title('A Graph Plot Showing Diabetes Progression Against BMI')
plt.show()
```
